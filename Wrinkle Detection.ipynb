# Problem Statement: Detect the Wrinkle around Eyes

## Image 1

```python
import cv2
import numpy as np
from IPython.display import display, Image

image = cv2.imread("laugh-lines-eyes.jpg")
display(Image(filename='laugh-lines-eyes.jpg'))
gray_img = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
eye_cascade = cv2.CascadeClassifier('haarcascade_eye.xml')
eyes = eye_cascade.detectMultiScale(gray_img, scaleFactor = 1.2,minNeighbors = 1)

for x,y,w,h in eyes: 
    cropped_img = gray_img[y+250:y+h-100,x:x+w+60]
    edges = cv2.Canny(cropped_img,200,250)        
    number_of_edges = np.count_nonzero(edges)
    
print(f'Number of Edges = {number_of_edges}')

if len(eyes)==2 and number_of_edges > 100:
    print("Wrinkle Found ")
elif len(eyes)==1 and number_of_edges > 50:
    print("Wrinkle Found ")
else:
    print("Wrinkle's NOT FOUND!")
    
for (x,y,w,h) in eyes:
        cv2.rectangle(image,(x,y+250),(x+w+60,y+h-100),(0, 255, 0),2)
        
cv2.imshow("Eyes Detected", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
